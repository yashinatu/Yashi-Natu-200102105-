```python
df=a.copy()
```

```python
df=df.rename({"Segment type": "stype"}, axis=1)
df=df.rename({"It became a relationship": "Result"}, axis=1)
```

```python
df
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
def f(s):
    if(s=="Yes"):
        return 1
    else:
        return 0

df.Answer = df.Answer.apply(f)
```

```python
def f(s):
    if(s=="web"):
        return 0
    elif(s=="gender"):
        return 1
    elif(s=="mobile"):
        return 2
    else:
        return 3

df.stype = df.stype.apply(f)
```

```python
df
```

```python
df.drop(["ID","Segment Description"],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size = 0.30)
```

```python
x=train.drop("Result",axis=1)
y=train.Result

x1=test.drop("Result",axis=1)
y1=test.Result
```

```python
model = LogisticRegression()
model.fit(x,y)
```

```python
prediction = model.predict(x1)
print("Accuracy from Logistic Regression model=",accuracy_score(y1,prediction))
```

```python
newmodel = LogisticRegression(random_state=220, C=3, multi_class="ovr").fit(x,y)
pre=newmodel.predict(x1)
print(accuracy_score(y1,pre))
```

Random Forest Classifier:

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model2=RandomForestClassifier().fit(x,y)
pre2=model2.predict(x1)

print("Accuracy of Random Forest Classifier = ",accuracy_score(y1,pre2))
```

#Changing Hyperparameters 

```python
newmodel2=RandomForestClassifier(n_estimators=50, max_depth=6, random_state=115).fit(x,y)
pre3=newmodel2.predict(x1)

print("On changing hyperparameters, accuracy = ",accuracy_score(y1,pre3))
```
